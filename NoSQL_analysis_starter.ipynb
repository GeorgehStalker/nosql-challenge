### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
print(establishments.count_documents(query))

# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
hygiene_more_than_20_df = pd.DataFrame(list(establishments.find(query)))

# Display the number of rows in the DataFrame
print(len(hygiene_more_than_20_df))
print(hygiene_more_than_20_df.shape)
print(hygiene_more_than_20_df.info())
```

```python
# Display the first 10 rows of the DataFrame
hygiene_more_than_20_df.head(10)
```

```python
# Provide the answer to question 1
print("The establishments with a hygiene score of 20 are:")
print(hygiene_more_than_20_df['BusinessName'])
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
            "LocalAuthorityName": {"$regex": "London"}, 
            "RatingValue": {"$gte": 4}
        }

# Use count_documents to display the number of documents in the result
print(establishments.count_documents(query))

# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
london_rating_more_than__df = pd.DataFrame(list(establishments.find(query)))

# Display the number of rows in the DataFrame
print(len(london_rating_more_than__df))

# Display the first 10 rows of the DataFrame
london_rating_more_than__df.head(10)
```

```python
# Provide the answer to question 2
print("The establishments in London with a RatingValue greater than or equal to 4 are:")
print(london_rating_more_than__df['BusinessName'])
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

# Define the search parameters
degree_search = 0.01

# Get the latitude and longitude of "Penang Flavours" from NoSQL database
query = {"BusinessName": "Penang Flavours"}
fields =  {"geocode.latitude": 1, "geocode.longitude": 1}

coords = establishments.find_one(query, fields)

latitude = coords.get('geocode').get('latitude')
longitude = coords.get('geocode').get('longitude')

# Find the top 5 establishments as instructed
query = {
    "geocode.latitude": {"$lte": latitude + degree_search, "$gte": latitude - degree_search},
    "geocode.longitude": {"$lte": longitude + degree_search, "$gte": longitude - degree_search},
    "RatingValue": 5
}

sort_query = {"scores.Hygiene": -1}

# Print the results
results = establishments.find(query).sort(sort_query).limit(5)
for result in results:
    pprint(result)
```

```python
# Convert result to Pandas DataFrame
penang_flavours_competitors_df = pd.DataFrame(list(establishments.find(query).sort(sort_query).limit(5)))
penang_flavours_competitors_df
```

```python
# Provide the answer to question 3
print("The top 5 establishments with a hygiene score of 5 near Penang Flavours are:")
print(penang_flavours_competitors_df['BusinessName'])
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
match_query = {"$match": {"scores.Hygiene": 0}}
group_query = {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}}
sort_query = {"$sort": {"count": -1}}

# Define the pipeline
pipeline = [
    match_query,
    group_query,
    sort_query
]

# Execute the pipeline
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"Number of documents: {len(results)}")

# Print the first 10 results
for result in results[:10]:
    pprint(result)
```

```python
# Convert the result to a Pandas DataFrame
hygiene_score_0_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
hygiene_more_than_20_df.shape

# Display the first 10 rows of the DataFrame
hygiene_score_0_df.head(10)
```

```python
# Provide the answer to question 4
# How many establishments in each Local Authority area have a hygiene score of 0?
print("The total establishments per Local Authorities area that have a hygiene score of 0 are:")
hygiene_score_0_df
```
